```python
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline
import seaborn as sns
import pandas as pd
```

```python
import matplotlib
matplotlib.rc('xtick', labelsize=12) 
matplotlib.rc('ytick', labelsize=12) 
```

```python
A = [3,3,2,5,2]
B = [1,6,2,4,2]
C = [6,5,2,1,1]
```

```python
cols = ['0 goals','1 goal','2 goals', '3 goals', '4 goals']
```

```python
# dataframe of goals scored by each team
Alpha_United = pd.DataFrame(A,index=cols, columns = ['Alpha United'])
Beta_Rovers = pd.DataFrame(B,index=cols, columns = ['Beta Rovers'])
Camma_City = pd.DataFrame(C,index=cols, columns = ['Camma City'])
Goals = pd.concat([Alpha_United,Beta_Rovers,Camma_City], axis=1)
Goals
```

```python
# pie charts showing each team's goals with the labels positioned
def alpha_pie(axes, k = [.35,.4,1.05], data=Goals, alpha=.65):
    axes.set_title("Alpha United", size = 14)
    axes.set_aspect(1)
    patches, texts = axes.pie(data.iloc[:,0], labels = [data.index[0],"",data.index[2],data.index[3],data.index[4]], labeldistance = k[0], colors=colours,
                                   wedgeprops = {'edgecolor':'black', 'alpha':.5})
    plt.setp(texts,size=12)
    patches, texts = axes.pie(data.iloc[:,0], labels = ["",data.index[1],"", "",""], labeldistance = k[0]+.25, colors=colours,
             wedgeprops = {'edgecolor':'black', 'alpha':0})
    plt.setp(texts,size=12)
```

```python
def beta_pie(axes, k = [.35,.4,1.05], data=Goals, alpha=.65):
    axes.set_title("Beta Rovers", size = 14)
    axes.set_aspect(1)
    patches, texts = axes.pie(data.iloc[:,1], labels = [data.index[0],data.index[1],data.index[2],"",data.index[4]], labeldistance = k[1], colors=colours,
                                   wedgeprops = {'edgecolor':'black', 'alpha':.5})
    plt.setp(texts,size=12)
    patches, texts = axes.pie(data.iloc[:,1], labels = ["","","",data.index[3],""], labeldistance = k[1]+.2, colors=colours,
             wedgeprops = {'edgecolor':'black', 'alpha':0})
    plt.setp(texts,size=12)
```

```python
def camma_pie(axes, k = [.35,.4,1.05], data=Goals, alpha=.65):
    axes.set_title("Camma City", size = 14)
    axes.set_aspect(1)
    labs = list(Goals.index[:2])
    labs.append(" ")
    labs.append(" ")
    labs.append(" ")
    patches, texts = axes.pie(data.iloc[:,2], labels = labs, labeldistance = k[2-1], 
                                   colors=colours, wedgeprops = {'edgecolor':'black', 'alpha':alpha})
    plt.setp(texts,size=12)
    patches, texts = axes.pie(data.iloc[:,i], labels = ["","",data.index[2], "",""], labeldistance = k[i]+.05, colors=colours,
             wedgeprops = {'edgecolor':'black', 'alpha':0})
    plt.setp(texts,size=12)
    patches, texts = axes.pie(data.iloc[:,i], labels = ["","","",data.index[3], Goals.index[4]], labeldistance = k[i], colors=colours,
             wedgeprops = {'edgecolor':'black', 'alpha':0})
    plt.setp(texts,size=12)
```

```python
# bar charts to show each team's number of goals
def bars(i, axes, data=Goals, alpha = .65):
    axes.set_title(Goals.columns[i], size = 14)
    axes.set_ylim(bottom=0,top=6.5)
    a,b,c,d,e = axes.bar(np.arange(5),Goals.iloc[:,i], tick_label = Goals.index, alpha = alpha, edgecolor='black')
    a.set_facecolor('blue')
    b.set_facecolor('orange')
    c.set_facecolor('green')
    d.set_facecolor('red')
    e.set_facecolor('purple')
```

```python
# export a pie chart and bar chart for each team
k=[.35,.4,1.05]
colours = ['b','orange','green','red','purple']
chart = [[],[],[]]
ax=[[],[],[]]
for i in range (3):
    chart[i] = plt.figure (figsize = (5,10))
    ax[i] = chart[i].subplots(2,1)
    bars(i,ax[i][1])
alpha_pie(ax[0][0])
beta_pie(ax[1][0])
camma_pie(ax[2][0])

chart.savefig(str(Goals.columns[i]))
```
